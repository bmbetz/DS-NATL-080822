## Task 10: Bar Chart

Create a bar chart showing the number of posts per month.

```python
# Your code here
# First need to access the number of posts per month
fig, ax = plt.subplots(figsize=(8, 5))
x = df_facebook['Post Month'].value_counts().sort_index().index

y = df_facebook['Post Month'].value_counts().sort_index().values

ax.bar(x,y)
ax.set_xlabel('Month')
ax.set_ylabel('Number of Posts')
ax.set_title('Posts Per Month');
```
